```python
# Getting Dependent and Independent variable.
x=auto['engine-size']
y=auto['city-mpg']
```

```python
# importing the desired Library.
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# assigning the value to train test variables.
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
# storing linear Regression to a variable
lr=LinearRegression()
```

```python
# Training the data
lr.fit(np.array(x_train).reshape(-1,1),np.array(y_train).reshape(-1,1))
```

```python
# Testing the data
y_pred=lr.predict(np.array(x_test).reshape(-1,1))
```

```python
# importing the mean squared error method
from sklearn.metrics import mean_squared_error
```

```python
# find the mean square error.
mean_squared_error(x_test,y_pred)
```
